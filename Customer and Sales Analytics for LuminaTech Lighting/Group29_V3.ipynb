```python
import pandas as pd

def remove_outliers(df, column):
    # Calculate lower and upper bounds for each group
    bounds = df.groupby(['company_code', 'order_type_code', 'abc_class_volume','customer_district_code'])[column].agg(
        lower_bound=lambda x: x.quantile(0.25) - 1.5 * (x.quantile(0.75) - x.quantile(0.25)),
        upper_bound=lambda x: x.quantile(0.75) + 1.5 * (x.quantile(0.75) - x.quantile(0.25))
    ).reset_index()

    # Merge bounds back to the original DataFrame for filtering
    df_with_bounds = df.merge(bounds, on=['company_code', 'order_type_code', 'abc_class_volume','customer_district_code'])
    
    # Filter out rows outside the IQR bounds for the specified column
    df_no_outliers = df_with_bounds[(df_with_bounds[column] >= df_with_bounds['lower_bound']) & 
                                    (df_with_bounds[column] <= df_with_bounds['upper_bound'])]

    # Drop the temporary bounds columns
    df_no_outliers = df_no_outliers.drop(columns=['lower_bound', 'upper_bound'])
    
    return df_no_outliers

# Apply the function for each column
data_no_outliers = remove_outliers(data4, 'value_quantity')
data_no_outliers = remove_outliers(data_no_outliers, 'value_sales')
data_no_outliers = remove_outliers(data_no_outliers, 'value_cost')
```

### 3.1. Is there a significant difference between the average sales for Company 101 (company with the highest sales) between 2012 and 2013? 

We want to assess whether there is a significant difference in the average sales for Company 101 between 2012 and 2013.

Hypotheses:
Null Hypothesis (H0): There is no difference in average sales for company 101 between 2012 and 2013.
Alternative Hypothesis (H1): There is a significant difference in n average sales for company 101 between 2012 and 2013.

Statistical Test: a two-sample t-test to compare the means of value_sales between the two years.

Before runninng the two sample t-test, we will check to see which company has the highest sales across both year. 

```python
# Assuming data_no_outliers is your DataFrame
result = data_no_outliers.groupby('company_code', as_index=False)['value_sales'].sum()

# Sort the result by value_sales in descending order
result_sorted = result.sort_values(by='value_sales', ascending=False)

# Display the result
result_sorted
```

From the table above we can see that Company 101 has the most sales with 2012 and 2013 put together. However, we cannot tell if the sales for company 101 dropped or increased from 2012 to 2013.

```python
from scipy import stats

# Creating a new DataFrame to hold the results of sales per year (2012 and 2013) per company code 
sales_by_year = data_no_outliers.groupby('company_code').agg(
    value_sales_2012=('value_sales', lambda x: x[data_no_outliers['calendar_year'][x.index] == 2012].sum()), 
    value_sales_2013=('value_sales', lambda x: x[data_no_outliers['calendar_year'][x.index] == 2013].sum())
).reset_index()

# Filtering the sales_by_year DataFrame for Company 101
company_101_sales = sales_by_year[sales_by_year['company_code'] == '101']

# Printing the results for Company 101
print(company_101_sales)

# Conducting the two-sample t-test using the existing calendar_year column for Company 101
sales_2012 = data_no_outliers[(data_no_outliers['company_code'] == '101') & (data_no_outliers['calendar_year'] == 2012)]['value_sales']
sales_2013 = data_no_outliers[(data_no_outliers['company_code'] == '101') & (data_no_outliers['calendar_year'] == 2013)]['value_sales']

# Conducting the two-sample t-test
t_statistic, p_value = stats.ttest_ind(sales_2012, sales_2013)

# The results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Interpreting the result by comparing p-value to level of significance
alpha = 0.05  
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in average sales between 2012 and 2013 for Company 101.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in average sales between 2012 and 2013 for Company 101.")
```

**Analysis:**

**1. The T-Statistic:** -47.82 is a very high in absolute terms and the negative sign indicates that the average sales in 2013 were lower than in 2012. This large absolute value indicates a significant difference between the two sample means - the average sales of Company 101 in 2012 and 2013.

**2. The P-Value:** 0.00 is below the level of significance of 0.05, indicating that there is a statistically significant difference between the average sales for Company 101 in 2012 and in 2013. 

By rejecting the H0 we conclude that there is a statistically significant difference in average sales between the years 2012 and 2013 for Company 101. Though company 101 had the highest sales in both 2012 and 2013, there is clear indication that its performance dropped in 2013. 

As seen in section 2.3 above, Company 101 sales started dropping in Q3 by hitting an all time low in January 2013. Its sales did recover from January and saw a steady climb until Q3 of 2013; however, the sales did not reach the levels Company 101 saw in 2012.

### 3.2. Is There a Difference in Average Quantity Sold/Sales Value Between Trade and Professional Bonus Groups?

We want to assess whether there is a significant difference in the average quantity sold between the Trade and Professional bonus groups.

Hypotheses:
Null Hypothesis (H0): There is no difference in average quantity sold between the Trade and Professional bonus groups.
Alternative Hypothesis (H1): There is a difference in average quantity sold between the Trade and Professional bonus groups.

Statistical Test:

Perform a two-sample t-test to compare the means of value_quantity between the two groups.

```python
from scipy import stats
# Filtering the dataset for the Trade and Professional groups
trade_quantity = data_no_outliers[data_no_outliers['bonus_group_code'] == 'Trade']['value_quantity']
professional_quantity = data_no_outliers[data_no_outliers['bonus_group_code'] == 'Professional']['value_quantity']

# Performing the two-sample t-test
t_stat_quantity, p_value_quantity = stats.ttest_ind(trade_quantity, professional_quantity)

# Calculating means for insights
mean_trade_quantity = trade_quantity.mean()
mean_professional_quantity = professional_quantity.mean()

print(f"Mean Quantity Sold (Trade): {mean_trade_quantity}")
print(f"Mean Quantity Sold (Professional): {mean_professional_quantity}")
print(f"T-statistic: {t_stat_quantity}, P-value: {p_value_quantity}")
```

Interpretation:

The average quantity sold in the Professional group (10.73) is significantly higher than in the Trade group (9.23).
The t-statistic of -8.02 indicates that the means are quite far apart, and the negative sign shows that the Professional group's mean is higher.
The p-value is extremely low (much less than 0.05), leading to the rejection of the null hypothesis. This means there is a statistically significant difference in the average quantity sold between the two groups, with the Professional group selling more on average.
